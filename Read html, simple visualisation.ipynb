```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
url = "https://en.wikipedia.org/wiki/Demographics_of_Italy"
tables = pd.read_html(url, header=0)
table = tables[3]
```

```python
table.rename(columns={'Unnamed: 0' : 'Year'}, inplace=True)
table.set_value(116, 'Year', 2016)
table[['Year']] = table[['Year']].apply(pd.to_numeric)
table.head()
```

```python
table.plot(x='Year', y='Natural change')
```

```python
sns.lmplot('Year', 'Natural change', data=table, order=1)
sns.lmplot('Year', 'Natural change', data=table, order=2)
sns.lmplot('Year', 'Natural change', data=table, order=3)
sns.lmplot('Year', 'Natural change', data=table, order=4)
```
